```python
# TASK 3: Calculate the distances between a launch site to its proximities
import numpy as np
import pandas as pd

# Sample data (use your actual launch_sites_df)
# Assuming your DataFrame looks like this:
# launch_sites_df = pd.DataFrame({
#     'Launch Site': [...],
#     'Lat': [...],
#     'Long': [...]
# })

# Define Haversine function to calculate distance between two lat/lon points
def haversine(lat1, lon1, lat2, lon2):
    R = 6371  # Radius of Earth in kilometers
    phi1, phi2 = np.radians(lat1), np.radians(lat2)
    d_phi = np.radians(lat2 - lat1)
    d_lambda = np.radians(lon2 - lon1)

    a = np.sin(d_phi / 2.0)**2 + np.cos(phi1) * np.cos(phi2) * np.sin(d_lambda / 2.0)**2
    c = 2 * np.arcsin(np.sqrt(a))
    return R * c

# Proximity location example: Kennedy Space Center Visitor Complex
proximity_lat = 28.5721
proximity_lon = -80.6480

# Calculate distance and add to the DataFrame
launch_sites_df['Distance_to_Visitor_km'] = launch_sites_df.apply(
    lambda row: haversine(row['Lat'], row['Long'], proximity_lat, proximity_lon),
    axis=1
)

# Show updated DataFrame
print(launch_sites_df[['Launch Site', 'Distance_to_Visitor_km']])

```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```
